# Histogram/Distplot

```python
diamonds = pd.read_csv("diamonds.csv")
```

```python
diamonds.head()
```

```python
diamonds.head()
```

```python
sns.distplot(diamonds["price"])
plt.show()
```

```python
sns.distplot(diamonds["price"],hist=False)
plt.show()
```

```python
sns.distplot(diamonds["price"],kde=False) # Kde means frequency curve
plt.show()
```

```python
sns.distplot(diamonds["price"],color="blue")
plt.show()
```

```python
sns.distplot(diamonds["price"],color="green",bins=20,kde=False)
plt.show()
```

```python
sns.distplot(diamonds["price"],color="green",vertical=True)
plt.show()
```

# Box Plot

```python
churn = pd.read_csv("churn.csv")
```

```python
churn.head()
```

```python
churn.shape
```

```python
sns.boxplot(x="Churn",y="tenure",data=churn) # Tenure means for how many months Customer using your service
plt.show()                                   # Churn means Will Customer leave or not
```

```python
sns.boxplot(x="Churn",y="tenure",data=churn,palette="rocket")
plt.show()
```

```python
sns.boxplot(x="Churn",y="tenure",data=churn,palette="bright")
plt.show()
```

```python
sns.boxplot(x="InternetService",y="MonthlyCharges",data=churn)
plt.show()
```

```python
sns.boxplot(x="InternetService",y="MonthlyCharges",data=churn,palette="Set1")
plt.show()
```

```python
sns.boxplot(x="InternetService",y="tenure",data=churn,palette="Set2")
plt.show()
```

```python
sns.boxplot(x="InternetService",y="tenure",data=churn,color = "Blue")
plt.show()
```

```python
sns.boxplot(x="InternetService",y="tenure",data=churn,palette= "rocket",linewidth=4)
plt.show()
```

```python
sns.boxplot(x="Contract",y="tenure",data=churn)
plt.show()
```

```python
sns.boxplot(x="Contract",y="tenure",data=churn,order=["One year","Two year","Month-to-month"])
plt.show()
```

```python
sns.boxplot(x="Contract",y="tenure",data=churn,order=["One year","Two year"])
plt.show()
```

```python
churn.head()
```

```python
sns.boxplot(x="Contract",y="tenure",data=churn,hue="PaymentMethod")
plt.show()
```
