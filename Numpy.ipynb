### Joining Numpy Arrays

#### Vstack

```python
n12 = np.array([10,20,30])
n13 = np.array([40,50,60])
```

```python
np.vstack((n12,n13))
```

```python
n14 = np.array([[10,20,30],[15,25,35]])
n15 = np.array([[40,50,60],[45,55,65]])
```

```python
np.vstack((n14,n15))
```

#### hstack

```python
np.hstack((n12,n13))
```

```python
np.hstack((n14,n15))
```

#### Column_stack

```python
np.column_stack((n12,n13))
```

```python
np.column_stack((n14,n15))
```

### Numpy Intersection & Difference

```python
n16 = np.array([10,20,30,40,50,60])
n17 = np.array([50,60,70,80,90])
```

```python
np.intersect1d(n16,n17) # It will Help you to find the common elements in the list
```

```python
np.setdiff1d(n16,n17) # n16 - n17 
                      # Unique elements in n16
                      # find the set difference of two arrays and return the unique values in arr1 that are not in arr2
```

```python
np.setdiff1d(n17,n16) # Unique elements in n17 
```

```python
a = np.random.randint(0,100,50)
a
```

```python
b = np.random.randint(0,100,50)
b
```

```python
x = np.intersect1d(a,b)
x
```

### Numpy array Mathematics

```python
n18 = np.array([10,20,30,40])
n19 = np.array([50,60,70,80])
```

```python
np.sum([n18,n19])
```

```python
from IPython.display import Image
Image(filename='Axis.jpg')
```

```python
np.sum([n18,n19],axis=0) # Vertical Sum or Column sum
```

```python
np.sum([n18,n19],axis=1) # Horizontal Sum or Row sum
```

```python
n18 = np.array([[10,20,30,40],[10,60,70,80]])
n19 = np.array([[50,60,70,80],[10,20,30,40]])
```

```python
np.sum([n18,n19],axis= 2)
```
